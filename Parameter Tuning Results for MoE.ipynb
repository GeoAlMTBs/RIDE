```python
import matplotlib.pyplot as plt

description = '32 experts'
moe_top_k = [8, 16, 32]
acc = [0.4001, 0.3955, 0.4072]
plt.plot(moe_top_k, acc)
plt.xlabel('moe_top_k')
plt.ylabel('accuracy')
plt.title(description)
plt.show()
```
